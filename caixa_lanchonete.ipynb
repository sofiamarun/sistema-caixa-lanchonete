```python
import datetime

estoque = {"banana": [6.0, 36],
           "maçã": [5.0, 50],
           "laranja": [4, 70]}

def verificar_estoque(nome_do_produto):
  if nome_do_produto in estoque:
    return True
  else:
    return False

resposta = input("Deseja cadastrar um novo produto? (S/N) ")
while resposta.lower() == "s":
    nome = input("Digite o nome do produto: ")
    if verificar_estoque(nome):
      print("Produto já cadastrado.")
      resposta = input("Deseja cadastrar um novo produto? (S/N)")
    else:
      preco = float(input("Digite o preço do produto: "))
      qtd = int(input("Digite a quantidade do produto em estoque: "))
      estoque[nome] = [preco, qtd]
      print(estoque)
      print ("Produto cadastrado.")
      resposta = input("Deseja cadastrar um novo produto? (S/N)")

print("-------------------")
print("Cadastro encerrado.")

for i in estoque:
  print(i)
  print("Preço:", estoque[i][0])
  print("Quantidade:", estoque[i][1])

# consultar_produto = input("Digite qual produto deseja consultar ou 'fim', para encerrar o programa: ")
# while consultar_produto != "fim":
#   if consultar_produto in estoque:
#     print("Preço:", estoque[consultar_produto][0])
#     print("Quantidade:", estoque[consultar_produto][1])
#   else:
#     print("Produto indisponível.")
#   consultar_produto = input("Digite qual produto deseja consultar ou 'fim', para encerrar o programa: ")

def atualizar_historico(historico, venda):
  data = datetime.datetime.now()
  data_formatada = data.strftime("%H:%M:%S")
  print(f"Hora atual: {data_formatada}")
  historico[data_formatada] = venda

historico = {}
print(historico)
valor_total = 0
venda = {}
num_produtos = 0
resposta_compra = input("Digite o produto para adicionar ao carrinho ou 'encerrar' para encerrar a compra: ")
while resposta_compra != "encerrar":
  if verificar_estoque(resposta_compra):
    quantidade_compra = int(input(f"Digite a quantidade desejada de {resposta_compra}: "))
    valor_produto = estoque[resposta_compra][0]
    valor_total_produto = valor_produto * quantidade_compra
    valor_total = valor_total + (valor_produto * quantidade_compra)
    num_produtos += 1
    venda[f"produto {num_produtos}"] = [resposta_compra, quantidade_compra, valor_total_produto]
    resposta_compra = input("Digite o produto para adicionar ao carrinho ou 'encerrar' para encerrar a compra: ")
  else:
    print("Produto indisponível.")
    resposta_compra = input("Digite o produto para adicionar ao carrinho ou 'encerrar' para encerrar a compra: ")
print(f"Valor total da compra: R${valor_total}")
atualizar_historico(historico, venda)
print(historico)
```
